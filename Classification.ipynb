```python
import pandas as pd
```

## Data preview

```python
leaves_df = pd.read_csv("data/dataset.csv")
```

```python
leaves_df.head()
```

```python
leaves_df.describe()
```

## Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
features = list(leaves_df.iloc[:,:-1].columns)
x = leaves_df[features]
y = leaves_df["label"]
```

```python
std_scaler = StandardScaler()
std_scaler.fit(x)
```

```python
leaves_df_std = std_scaler.transform(x)
leaves_df_std = pd.DataFrame(leaves_df_std, columns=features)
leaves_df_std.head()
```

```python
leaves_df_std.describe()
```

## Model training

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
rdf = RandomForestClassifier(random_state=42)
rdf.fit(x_train, y_train)
```

## Model performance

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
y_pred = rdf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
